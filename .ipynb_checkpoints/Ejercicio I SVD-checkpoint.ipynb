### En una libreta de Jupyter, calcula la SVD de la matriz A = [[1, 2], [2, 3]]. Grafica el efecto sobre el conjunto de los vectores en R 2 de longitud 1 (el círculo de centro [0, 0] y radio 1) al aplicarle  VT, luego Σ y luego U. Compara el efecto de aplicarle al círculo directamente A.

```python
A = np.array([[1,2],[2,3]])
print(A)
```

```python
autovalores, autovectores = np.linalg.eig(A)
print(autovalores)
print(autovectores)
```

```python
U, D, V = np.linalg.svd(A)
```

```python
print(U) # matriz de U
```

```python
print(np.diag(D)) # matriz diagonal Σ ó D (se utiliza np.diag para que numpy coloque los ceros)
```

```python
print(V) # matriz VT
```

```python
#Graficando Matriz Original
print('Graficando A')
graficarMatriz(A)
plt.xlim(-2.5,2.5)
plt.ylim(-4,4)
plt.show()
```

```python
print('Graficando Efecto en V:')
graficarMatriz(V)
plt.xlim(-1.5,1.5)
plt.ylim(-1.5,1.5)
plt.show()
```

```python
print('Graficando Efecto en D :')
graficarMatriz(np.diag(D))
plt.xlim(-4.5,4.5)
plt.ylim(-1.5,1.5)
plt.show()
```

```python
print('Graficando Efecto Rotacion en D :')
graficarMatriz(np.diag(D).dot(V)) # Realizando producto punto de D y V
plt.xlim(-4.5,4.5)
plt.ylim(-1.5,1.5)
plt.show()
```

```python
print('Grqaficando efecto en U:')
graficarMatriz(U)
plt.xlim(-1.2,1.2)
plt.ylim(-4,4)
plt.show()
```

```python
print('Regresando a Matriz Original:')
graficarMatriz(U.dot(np.diag(D).dot(V))) #Producto Punto de U,D,V
plt.xlim(-2.5,2.5)
plt.ylim(-4,4)
plt.show()
```

#### Calcula la matriz de rank 1 que mejor aproxime a A

```python
u_rank1 = U[:,:1]  #obteniendo rank 1 de U (R1)
print(u_rank1)     
```

```python
d_rank1 = D[:1] #obteniendo rank 1 de D (R1)
print(d_rank1)
```

```python
v_rank1 = V[:1,:]  #obteniendo rank 1 de V (R1)
print(v_rank1)
```

```python
A_rank1 = u_rank1.dot(np.diag(d_rank1).dot(v_rank1)) #producto punto de los elementos con rank 1
print(A_rank1)
```

```python
print('Grafico Matriz de rank 1 de A:')
graficarMatriz(A_rank1)
plt.xlim(-2.5,2.5)
plt.ylim(-4,4)
plt.show()
```
